```python
import matplotlib.pyplot as plt
import json
```

```python
with open("rq2v4.json", "r") as f:
    data = json.load(f)
```

```python
keys = []
# deleted
deleted = []
# reenabled
reenabled = []
for k in data:
    keys.append(k.split("@")[0])
    deleted.append([t / 3600 / 24 for t in data[k]["durations"]["DELETED"]])
    reenabled.append([t / 3600 / 24 for t in data[k]["durations"]["RE-ENABLED"]])

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.boxplot(deleted)
plt.title("Duration from Disabled to Deleted")
plt.ylabel("days")
plt.xticks([i + 1 for i in range(15)], keys, rotation = 90)
plt.subplot(1, 2, 2)
plt.boxplot(reenabled)
plt.title("Duration from Disabled to Re-enabled")
plt.ylabel("days")
plt.xticks([i + 1 for i in range(15)], keys, rotation = 90);
```

```python
keys = []
# deleted
deleted = []
# reenabled
reenabled = []
for k in data:
    keys.append(k.split("@")[0])
    deleted.append([t for t in data[k]["commits"]["DELETED"]])
    reenabled.append([t for t in data[k]["commits"]["RE-ENABLED"]])

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.boxplot(deleted)
plt.title("Commits between Disabled to Deleted")
plt.xticks([i + 1 for i in range(15)], keys, rotation = 90)
plt.subplot(1, 2, 2)
plt.boxplot(reenabled)
plt.title("Commits between Disabled to Re-enabled")
plt.xticks([i + 1 for i in range(15)], keys, rotation = 90);
```
